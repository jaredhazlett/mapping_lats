```python
import pandas as pd
import requests
import json
from api_key_census import api_key
```

```python
variables = ("B25075_001E,B25075_015E,B25075_016E,B25075_017E,B25075_018E,B25075_019E,B25075_020E")
```

```python
housing_ca = requests.get(f"https://api.census.gov/data/2019/acs/acs5?get=NAME,{variables}&for=county:*&in=state:06&key={api_key}").json()
```

```python
housing_prices_CA = pd.DataFrame(housing_ca)
housing_prices_CA
```

```python
housing_prices_CA.columns = housing_prices_CA.iloc[0]
housing_prices_CA = housing_prices_CA[1:]
```

```python
#this is for later - our fips to english translation variable.
housing_fips_name = housing_prices_CA[["county", "NAME"]]
```

```python
housing_prices_CA = housing_prices_CA.drop(columns=["NAME"])
```

```python
housing_prices_CA = housing_prices_CA.apply(pd.to_numeric)
```

```python
housing_prices_CA = housing_prices_CA.rename(columns={
    "B25075_001E": "Total",
    "B25075_015E": "100k-124k",
    "B25075_016E": "125k-149k",
    "B25075_017E": "150k-174k",    
    "B25075_018E": "175k-199k",
    "B25075_019E": "200k-249k",  
    "B25075_020E": "250k-299k",
})
housing_prices_CA = housing_prices_CA.drop(columns=["state"])
housing_prices_CA
```

```python
housing_prices_CA["%100k-124k"] = housing_prices_CA["100k-124k"]/housing_prices_CA["Total"]*100
housing_prices_CA["%125k-149k"] = housing_prices_CA["125k-149k"]/housing_prices_CA["Total"]*100
housing_prices_CA["%150k-174k"] = housing_prices_CA["150k-174k"]/housing_prices_CA["Total"]*100
housing_prices_CA["%175k-199k"] = housing_prices_CA["175k-199k"]/housing_prices_CA["Total"]*100
housing_prices_CA["%200k-249k"] = housing_prices_CA["200k-249k"]/housing_prices_CA["Total"]*100
housing_prices_CA["%250k-299k"] = housing_prices_CA["250k-299k"]/housing_prices_CA["Total"]*100
```

```python
#we're only keeping the fraction of 
housing_prices_CA = housing_prices_CA.drop(columns=["Total", "100k-124k", "125k-149k", "150k-174k", "175k-199k", "200k-249k", "250k-299k"])
housing_prices_CA
```

```python
import geopandas as gpd
```

```python
#reading the shapefile with geopandas and filtering for counties with the "Caliornia" identifier
counties = gpd.read_file("cb_2020_us_county_5m/cb_2020_us_county_5m.shp")
counties = counties.loc[counties["STATE_NAME"] == "California"]
counties = counties[["COUNTYFP", "geometry"]]
counties
```

```python
chipotles = pd.read_csv("chipotle_stores.csv")
chipotles = chipotles.loc[chipotles["state"] == "California"]
chipotles = chipotles.drop(columns=["location", "address"])
```

```python
chipotles_gdf = gpd.GeoDataFrame(
    chipotles, geometry=gpd.points_from_xy(chipotles.longitude, chipotles.latitude), crs=4269)
```

```python
chipotles_gdf = chipotles_gdf.drop(columns=["latitude", "longitude", "state"])
```

```python
chipotle_map = gpd.sjoin(counties, chipotles_gdf, how='right')
chipotle_map = chipotle_map.drop(columns=["index_left"])
chipotle_map
```

```python
chipotle_map = chipotle_map.dropna(how='any')
chipotle_county_count = chipotle_map.drop(columns=["geometry"])
chipotle_county_count = chipotle_county_count.groupby("COUNTYFP").sum()
chipotle_county_count = chipotle_county_count.apply(pd.to_numeric)
```

```python
chipotle_county_count.dtypes
```

```python
chipotle_map["COUNTYFP"] = pd.to_numeric(chipotle_map["COUNTYFP"])
```

```python
chipotle_map
```

```python
chipotle_map_wealth = chipotle_map.merge(housing_prices_CA, left_on="COUNTYFP", right_on="county")
chipotle_map_wealth = chipotle_map_wealth.drop(columns=["COUNTYFP", "count", "geometry"])
chipotle_map_wealth
#chipotle_map_wealth.to_file("chipotle_map_wealth.geojson", driver='GeoJSON')
```

```python
chipotle_map_wealth = chipotle_map_wealth.drop_duplicates()
```

```python
chipotle_map_wealth = chipotle_map_wealth.apply(pd.to_numeric)
chipotle_map_wealth.dtypes
```

```python
chipotle_map_wealth = chipotle_map_wealth.merge(chipotle_county_count, how='inner', left_on='county', right_on='COUNTYFP')
chipotle_map_wealth
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(chipotle_map_wealth["%100k-124k"], chipotle_map_wealth["count"])
plt.title("Count of Chipotle by %100k-124k Valued houses")
plt.xlabel("%100k-124k")
plt.ylabel("Count of Chipotles")
```

```python
import statsmodels.api as sm
```

```python
X = chipotle_map_wealth.drop(columns=["county", "count"])
y = chipotle_map_wealth[["count"]]
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif["features"] = X.columns
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif.round(1)
#massive colinearities mean we need to separate out the data. These may be too close in housing costs.
```

```python
from sklearn.linear_model import LinearRegression
L = LinearRegression()
L.fit(X,y)
print("Coefficients: ",L.coef_)
print("Intercepts: ",L.intercept_)
```

```python
X = sm.add_constant(X)
```

```python
model = sm.OLS(y,X).fit()
```

```python
model.summary()
```

We can't tell based on the massive multicolinearities if chipotles increase/decrease from less wealthy neighborhoods to wealthy neighborhoods. Our P-values are not statistically significant. We should separate out the house wealth to a greater degree if we want to find any kind of pattern.
